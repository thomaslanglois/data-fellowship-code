# Data cleaning

Things to be dropped from the dataframe:
> 1. NaT indexes
> 2. columns with NaN values

```python
# This drops NaT indexes and then drops columns (axis=1) with NaN values
df = df.loc[df.index.dropna()].dropna(axis=1)
# We make sure that the index is of datetime type
df.index = pd.to_datetime(df.index)
```

```python
df.head(5)
```

```python
n_top_borrows = 5
top_borrows = df.loc[[df.index.min(), df.index.max()]].pct_change().dropna() \
        .transpose() \
        .sort_values('2018-12-01', ascending=False) \
        .head(n_top_borrows).index.values
```
