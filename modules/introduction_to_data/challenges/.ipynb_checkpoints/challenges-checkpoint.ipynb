## Challenge 2 - Cleaning data

### Cleaning the `happiness` database

```python
df_csv.head()
```

```python
df_csv = df_csv.rename(columns={'C': 'Country', 'R': 'Region', 'He': 'Hemisphere', 'Ha': 'HappinessScore', 'GDP': 'GDP_PerCapita', 'B': 'Beer_PerCapita', 'S': 'Spirit_PerCapita', 'W': 'Wine_PerCapita'})
```

```python
df_csv.head()
```

```python
df_csv.dtypes
```

```python
df_csv.HappinessScore = df_csv.HappinessScore.str.replace(',','.').astype('float64')
df_csv.GDP_PerCapita = df_csv.GDP_PerCapita.str.replace(',','.').astype('float64')
```

```python
df_csv.dtypes
```

```python
df_csv.Hemisphere.unique()
```

```python
df_csv.Hemisphere = df_csv.Hemisphere.replace(['North', 'south', 'N', 'north', 'noth',
                                               'nort', 'n', 'both', 's', 'S', 'South',
                                               'sout', 'southh'],
                                              ['north', 'south', 'north', 'north', 'north',
                                               'north', 'north', 'both', 'south', 'south', 'south',
                                               'south', 'south'])
```

```python
df_csv.Hemisphere.unique()
```

### Cleaning the `gapminder_data` database

```python
df_gm = pd.read_excel('C:\\Users\\thomas\\Documents\\Dev\\ml\\data-fellowship-code\\modules\\introduction_to_data\\datasets\\indicator_t 25-54 unemploy.xlsx', sheet_name='Data')
```

```python
df_gm.head()
```

```python
df_gm = df_gm.rename(columns={df_gm.columns.values[0]: 'COUNTRY'})
```

```python
df_gm.head()
```

```python
gmd_melt = pd.melt(df_gm, id_vars= ['COUNTRY'])
gmd_melt = gmd_melt.rename(columns={'variable': 'year', 'value': '25_54_unemployment'})
gmd_melt.head()
```

```python
gmd_melt.dtypes
```

```python
gmd_melt.year = gmd_melt.year.astype('int64')
gmd_melt.dtypes
```

```python
gmd_noNaN = gmd_melt.dropna()
```

```python
print('We dropped {} rows.'.format(gmd_melt.shape[0] - gmd_noNaN.shape[0]))
```

## Challenge 3 - Summarising data

```python
totalCounts = df_csv.Country.count()
totalCounts
```

```python
df_csv.head()
```

```python
summation = df_csv.HappinessScore.sum()
summation
```

```python
mean = summation / totalCounts
mean
```

```python
(mean, median, mode) = (df_csv.HappinessScore.mean(),
                        df_csv.HappinessScore.median(),
                        df_csv.HappinessScore.mode()[0])
print('HappinessScore:\n\tmean: {}\n\tmedian: {}\n\tmode: {}'.format(mean, median, mode))
```

```python
(var,std, q1, q3) = (df_csv.HappinessScore.var(),
                     df_csv.HappinessScore.std(),
                     df_csv.HappinessScore.quantile(.25),
                     df_csv.HappinessScore.quantile(.75))
print('HappinessScore:\n\tvariance: {}\n\tstd deviation: {}\n\tIQR: ({},{})'.format(var, std, q1, q3))
```

```python
(maxv,minv) = (df_csv.HappinessScore.max(),
               df_csv.HappinessScore.min())
print('HappinessScore:\n\tmax: {}\n\tmin: {}'.format(maxv, minv))
```

## Challenge 4 - Summarising data visually

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.boxplot(data=df_csv[['Beer_PerCapita','Spirit_PerCapita', 'Wine_PerCapita']])
plt.title('Country Alcohol Consumption')
plt.ylabel('Litres per capita (l)')
plt.xticks(ticks=(0, 1, 2), labels=['Beer', 'Spirits', 'Wine'])
plt.show()
```

```python
sns.boxplot(y='Hemisphere', x='HappinessScore', data=df_csv, 
            order=['north', 'both', 'south'])
plt.title('Country Happiness Score')
plt.xlabel('Happiness Score')
plt.ylabel('Hemisphere')
plt.yticks(ticks=(0, 1, 2), labels=['North', 'Both', 'South'])
plt.show()
```

```python
sns.violinplot(x='Hemisphere', y='GDP_PerCapita',
               order=('north', 'both', 'south'), data=df_csv)
plt.title('Gross domestic product per capita')
plt.ylabel('GDP per capita')
plt.xlabel('Region')
plt.xticks(ticks=(0, 1, 2), labels=('North', 'Both', 'South'))
plt.show()
```

```python
sns.scatterplot(x='Wine_PerCapita', y='HappinessScore', hue='Hemisphere', data=df_csv)
plt.title('Wine per capita and happiness')
plt.ylabel('Happiness score')
plt.xlabel('Litres of wine per capita (l)')
plt.show()
```

```python
sns.distplot(df_csv['HappinessScore'], kde=False, bins=10)
plt.title('Happiness Score Distribution')
plt.ylabel('Frequency')
plt.xlabel('Happiness score')
plt.show()
```

```python
sns.distplot(df_csv['HappinessScore'], hist=False)
plt.title('Happiness Score Distribution')
plt.ylabel('Density')
plt.xlabel('Happiness score')
plt.show()
```

```python
sns.pairplot(df_csv, hue='Hemisphere')
```
