```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier,export_graphviz
import graphviz
```

```python
data_mariak = pd.read_csv('./datasets/hr_mariak.csv')
data_full = pd.read_csv('./datasets/hr_full.csv')
data_mariak.head()
```

```python
data_full.head()
```

```python
data_full.shape
```

```python
data_mariak['hourly_salary'] = data_mariak['salary'] / data_mariak['average_monthly_hours']
data_full['hourly_salary'] = data_full['salary'] / data_full['average_monthly_hours']
data_mariak.head()
```

```python
train_x = data_full.drop(columns=['resigned'])
train_y = data_full['resigned']
test_x = data_mariak.drop(columns=['resigned'])
labels = list(train_x.columns)
```

```python
train_x.head()
```

```python
test_x
```

```python
labels
```

```python
clf = DecisionTreeClassifier()
clf.fit(train_x,train_y)
# Visualize the tree
from IPython.display import display
display(graphviz.Source(export_graphviz(clf, rounded=True, filled=True, class_names=['employed', 'resigned'], feature_names=labels)))
```

```python
prediction = clf.predict(test_x)
print('Will Maria K. resign? Prediction: {}'.format(bool(prediction)))
```
