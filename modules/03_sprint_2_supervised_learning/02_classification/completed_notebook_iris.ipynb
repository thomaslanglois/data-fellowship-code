```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
data['target'] = data['target'].replace([0,1,2],['Iris setosa', 'Iris versicolor', 'Iris virginica'])
```

```python
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
pair = sns.pairplot(data, hue="target")
pair
# Run this cell twice if you have a warning message to see the plot
```

```python
f, axes = plt.subplots(2, 2, figsize=(17, 17))
sns.boxplot(x = 'target', y = 'sepal length (cm)',data=data, ax=axes[0, 0])
sns.boxplot(x = 'target', y = 'sepal width (cm)',data=data, ax=axes[0, 1])
sns.boxplot(x = 'target', y = 'petal length (cm)',data=data, ax=axes[1, 0])
sns.boxplot(x = 'target', y = 'petal width (cm)',data=data, ax=axes[1, 1])
```
