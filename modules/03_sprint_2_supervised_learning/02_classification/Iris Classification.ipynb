# Classification case study: IRIS

## IRIS dataset

In 1936, the first dataset for statistical classification was made publicly available and has become the de facto example of why classification algorithms can be important.

The dataset consists of 50 samples of three species of the Iris plant; Iris setosa, Iris virginica and Iris veriscolor. For each sample, there were four measurements taken off the plants; length and width of the sepals and petals.

In theory, using these four measurements it is possible to identify (statistically) the species identity of any Iris plant collected in Gasp√© Peninsula, Canada. In other words, it is possible to label unlabelled data based on previously labelled data.

Let's have a look at the data:

# Import librairies

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load the data

```python
df = pd.read_csv('./datasets/iris.csv')
df.shape
```

```python
df.head()
```

## Understanding the dataset

We can start understanding our data better by plotting our variables on a box plot, independently of knowing the Iris species. This will give us information about the variables that are in the dataset and their distribution. It is also a good way to start thinking about outliers. 

```python
sns.set(style="ticks",font_scale=1.7)
plt.figure(figsize=(20, 10))
sns.boxplot(data=df)
```

For instance, for **sepal width**, we can see that there is a tight grouping in the middle of the range with few points sitting far outside the others.

On the other hand, **petal width** seems to have a large range of possible values. When the range of possible values is larger, it may indicate that a classification model of Iris species can break that large variable range, but we need to test that.

## Class Comparison

It's often the case that some of these properties may not tell us as much as others about which class a specific sample belongs to. To help identify which are the most useful, we can plot distributions for each feature and colour them by the class they belong to.

This also lets us see any correlation between the variables.

```python
sns.set(style="ticks",font_scale=1.2)
sns.pairplot(df, hue="species")
```

This visualisation makes it clearer which variables are most helpful in identifying a specific sample's class.

For example, I would guess **petal length** and **petal width** are great indicators of which class a sample belongs to, as the overlap across species is minimal. In addition, **sepal width** seems to be less informative as the overlap is so high.

However, it is always important to quantify the results to clearly understand the differences between Iris species.

## Further Analysis

Hopefully you can see now that, by plotting simple visualisations, we can get an understanding of which variables are going to be important in classifying the different species of Iris. Let's plot those box plots again, but this time separate them out for each species of Iris

```python
f, axes = plt.subplots(2, 2, figsize=(17, 17))
sns.boxplot(x='species', y='sepal_length', data=df, ax=axes[0, 0])
sns.boxplot(x='species', y='sepal_width', data=df, ax=axes[0, 1])
sns.boxplot(x='species', y='petal_length', data=df, ax=axes[1, 0])
sns.boxplot(x='species', y='petal_width', data=df, ax=axes[1, 1])
```

Wow!

So as we guessed, **petal length** definitely differs for each type, and even **petal width** (although less obvious) differs for each type.

**Sepal length** may give us an indication, but **sepal width** looks like it could get us confused if we tried to use it for classification.

This is the first step to any classification problem ‚Äì understanding what features are useful to classify the dataset. 

Sometimes it is not possible to know that because the dataset is too large (with too many variables to visualise all of them) and the labels are numerous (in this case there are only three labels). In these cases, a good classification model will help us to know which variables are contributing more to separate groups and to determine the logic rules for each group.

## Conclusion

In our analysis above, we did not use any classification algorithms. Instead, we used basic visualisation techniques. Visualising your data is always a good practice, as it helps us as data analysts get a good understanding of the dataset before we begin an exhaustive analysis.
